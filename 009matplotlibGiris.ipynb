```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
%matplotlib inline # bizim için direk grafiği verecek
```

```python
yasListesi =[10,20,30,40,50]
```

```python
kiloListesi =[55,66,33,66,32]
```

```python
numpyYasList= np.array(yasListesi)
numpyKiloList = np.array(kiloListesi)
```

```python
numpyKiloList
```

```python
plt.plot(numpyYasList,numpyKiloList,"g") #g green demek ilki X ekseni diğeri Y
plt.xlabel("Yas")
plt.ylabel("kilo")
```

```python
numpy1= np.linspace(0,10,20)
```

```python
numpy1
```

```python
numpy2 = numpy1 ** 3
```

```python
numpy2
```

```python
plt.plot(numpy1,numpy2,"g--") #g*- yaparsan farklı çizgi olr
```

```python
plt.subplot(1,2,1)
plt.plot(numpy1,numpy2,"g*-")
plt.subplot(1,2,2) # 1 sıra olacak 2 kolon olacak 2.çiziyorum
plt.plot(numpy1,numpy2,"b--")
```

```python
bfigure = plt.figure()
figureAxes = bfigure.add_axes([0.2,0.2,0.9,0.9]) #0.9 lar yükseklkgenişlik
figureAxes.plot(numpy1,numpy2,"b")
figureAxes.set_xlabel("x ekseni")
figureAxes.set_ylabel("y ekseni")
figureAxes.set_title("Garfik başlığı")
```

```python

figure2=plt.figure()

eksen1=figure2.add_axes([0.1,0.1,0.7,0.7])
eksen2=figure2.add_axes([0.3,0.3,0.3,0.3]) #0.3 ler ilk baştaki başlayıp bitme noktsaı grafiklerin


eksen1.plot(numpy1,numpy2,"g")
eksen1.set_xlabel("x eksenşi")
eksen1.set_ylabel("y ekseni")
eksen1.set_label("Ana Grafik Başlığı")


eksen1.plot(numpy2,numpy1,"y")
eksen1.set_xlabel("x eksenşi")
eksen1.set_ylabel("y ekseni")
eksen1.set_label("Küçük Grafik Başlığı")
```

```python
(benimFigure,benimEksen)=plt.subplots(nrows=1,ncols=2)

for eksen in benimEksen:
    eksen.plot(numpy1,numpy2,"g")
    eksen.set_xlabel("x ekseni")
plt.tight_layout()#iki grafik arası uzaklıgı ayarlar
```

```python
yeniFigur =plt.figure(dpi=100) #görüntü kalitesi diyebiliriz
yenieksen=yeniFigur.add_axes([0.2,0.3,0.9,0.9])
yenieksen.plot(numpy1,numpy2,label="numpyDizisi**2")
yenieksen.plot(numpy1,numpy2**3,label="nmp ** 3")
yenieksen.legend(loc=5)  # sol üst sağ alt vs yerini değiştirir yazıların
```

```python
yeniFigur.savefig("benimFigur.png",dpi=100)
```
